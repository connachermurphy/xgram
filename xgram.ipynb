```python
import torch
import matplotlib.pyplot as plt

from xgram import names_data
```

```python
# Let's count bigrams
names = names_data.names

bigrams_dict = {}

for n in names:
    chs = ["<S>"] + list(n) + ["<E>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        bigrams_dict[bigram] = bigrams_dict.get(bigram, 0) + 1
```

```python
sorted(bigrams_dict.items(), key = lambda x: x[1], reverse = True)
```

```python
sorted(bigrams_dict.items(), key = lambda x: x[1], reverse = False)
```

```python
chars = sorted(list(set("".join(names))))
chtoi = {ch: i + 1 for i, ch in enumerate(chars)}
chtoi["."] = 0

itoch = {i: ch for ch, i in chtoi.items()}
```

```python
chtoi
```

```python
itoch
```

```python
# Cast as a 2D array
N = torch.zeros((len(chtoi), len(chtoi)), dtype = torch.int32)

for n in names:
    chs = ["."] + list(n) + ["."]
    for ch1, ch2 in zip(chs, chs[1:]):
        ich1 = chtoi[ch1]
        ich2 = chtoi[ch2]
        N[ich1, ich2] += 1
```

```python
%matplotlib inline

plt.figure(figsize=(18, 18))
plt.imshow(N, cmap="Greens")

for i in range(len(itoch)):
    for j in range(len(itoch)):
        chs = itoch[i] + itoch[j]

        plt.text(j, i, N[i, j].item(), ha="center", va="top", color="gray")
        plt.text(j, i, chs, ha="center", va="bottom", color="gray")

plt.axis("off")
plt.tight_layout()
plt.savefig("bigram_matrix.png")
plt.show()
```
